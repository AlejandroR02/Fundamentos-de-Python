# Conceptos sobre grafos

Un grafo es un conjunto de nodos (vértices) y enlaces (vínculos, aristas o enlaces).

# Creación y modificación de grafos simples

## Crear un grafo circular

```python
# Importar las librerias requeridas
import matplotlib.pyplot as plt
import networkx as nx

# Crear un grafo con 12 nodos
grafo_0 = nx.cycle_graph(12)

# Mostrar el dibujo del grafo
plt.figure(figsize=(5, 5))
nx.draw(grafo_0, with_labels=True, font_size=12)
plt.show()

# Crear la matriz de adyacencia del grafo
matrix_adj_0 = nx.adjacency_matrix(grafo_0)

# Mostrar la matriz de adyacencia
print("Este es el contenido de la matriz de adyacencia:\n",
      matrix_adj_0.todense(), "\n")

# Mostrar los elementos activos de la matriz de adyacencia
print("Estos son los elementos activos de la matriz de adyacencia:\n",
      matrix_adj_0, "\n")

# Mostrar los nodos activos del grafo
print("Este es el contenido de los nodos activos del grafo:\n", grafo_0.nodes,
      "\n")

# Mostrar una posición específica de la matriz de adyacencia
print("Esta es la posición [1,2] de la matriz de adyacencia:\n",
      matrix_adj_0[1, 2])
```

## Crear un grafo vacío

```python
# Importar las librerias requeridas
import matplotlib.pyplot as plt
import networkx as nx

# Crear un grafo vacío
grafo_1 = nx.Graph()

# Mostrar el grafo
print("Este es el contenido del grafo_1:\n", grafo_1)

# Mostrar el dibujo del grafo
nx.draw(grafo_1, with_labels=True, font_size=12)
plt.show()
```

## Agregar elementos a un grafo

### Agregar nodos

```python
# Agregar un nodo
grafo_1.add_node(0)

# Mostrar el grafo
print("Este es el contenido del grafo_1:\n", grafo_1, "\n")

# Agregar varios nodos desde una lista
grafo_1.add_nodes_from([1, 2, 3])

# Mostrar el grafo
print("Este es el contenido del grafo_1:\n", grafo_1, "\n")

# Agregar nodos desde un objeto iterable
grafo_1.add_nodes_from(range(4, 7))

# Mostrar el contenido del grafo
print("Este es el contenido del grafo_1:\n", grafo_1, "\n")

# Mostrar el dibujo del grafo
nx.draw(grafo_1, with_labels=True, font_size=12, node_color="red")
plt.show()
```

### Agregar enlaces

```python
# Agregar un enlace
grafo_1.add_edge(1, 2)

# Mostrar el grafo
print("Este es el contenido del grafo_1:\n", grafo_1, "\n")

# Agregar varios enlaces desde un objeto iterable
grafo_1.add_edges_from([(2, 3),(2, 5), (3, 4)])

# Mostrar el contenido del grafo
print("Este es el contenido del grafo_1:\n", grafo_1, "\n")

# Mostrar el dibujo del grafo
plt.figure(figsize=(5, 5))
#pos = nx.random_layout(grafo_1)
nx.draw(grafo_1, with_labels=True, font_size=12,
        node_color="green", node_size = 500)
#nx.draw_networkx_labels(grafo_1, pos)
plt.show()
```

## Características de nodos y enlaces

```python
# Mostrar la información del grafo
# print("Este es el contenido del grafo_1:\n", nx.info(grafo_1), "\n")

# Obtener un objeto iterable con los nodos del grafo
print("Los nodos del grafo son:\n", nx.nodes(grafo_1), "\n")

print("Otra forma de obtener los nodos del grafo:\n", list(grafo_1.nodes),"\n")

for i in list(grafo_1.nodes):
  print(i)

# Mostrar el número de nodos del grafo
print("Este es el número de nodos del grafo:\n",
      nx.number_of_nodes(grafo_1), "\n")
print("Otra forma de obtener el número de nodos del grafo:\n",
      len(grafo_1.nodes), "\n")
```

## Agregar un grafo a  otro grafo

```python
# Crear un grafo
grafo_1 = nx.path_graph(10)

# Agregar un grafo a otro
grafo_0.add_nodes_from(grafo_1)

# Mostrar el grafo
print("Este es el contenido del grafo_0:\n", grafo_0, "\n")

# Mostrar el dibujo del grafo
nx.draw(grafo_0, with_labels=True, font_size=12)
plt.show()
```

## Agregar un grafo como un nodo a otro grafo

```python
# Agregar un grafo como un nodo a otro grafo
grafo_0.add_node(grafo_1)

# Mostrar el grafo
print("Este es el contenido del grafo_0:\n", grafo_0, "\n")

# Mostrar el dibujo del grafo
nx.draw(grafo_0, with_labels=True, font_size=12)
plt.show()
```

## Examinar los elementos de un grafo

* Listado de nodos
* Listado de enlaces
* Adyacencia
* Grado

```python
# Número de nodos de un grafo
print("El número de nodos del grafo:\n", grafo_0.number_of_nodes())

# Número de enlaces de un grafo
print("El número de enlaces del grafo:\n",grafo_0.number_of_edges())
```

## Eliminar elementos de un grafo

```python
# Eliminar un nodo
grafo_0.remove_node(11)

# Eliminar varios nodos
grafo_0.remove_nodes_from([2, 3])

# Eliminar un enlace
grafo_0.remove_edge(4, 5)

# Eliminar varios enlaces
grafo_0.remove_edges_from([(1, 2), (2, 3)])

# Vaciar el contenido de un grafo
grafo_0.clear()
```

# Análisis de gráfos

## Crear un grafo para hacer los analisis

```python
# Importar las librerías requeridas
import matplotlib.pyplot as plt
import networkx as nx

# Crear un grafo en que todos los nodos están conectados entre sí
grafo_ejemplo = nx.complete_graph(8)

# Mostrar el dibujo del grafo
plt.figure(figsize=(5, 5))
nx.draw(grafo_ejemplo, with_labels=True, font_size=12)
plt.show()

# Crear la matriz de adyacencia del grafo
matrix_adj_ejemplo = nx.adjacency_matrix(grafo_ejemplo)

# Mostrar la matriz de adyacencia
print("Este es el contenido de la matriz de adyacencia:\n",
      matrix_adj_ejemplo.todense(), "\n")
```

## Modificar el grafo para hacer los analisis

```python
# Hacer una copia del grafo ejemplo
grafo_ejemplo_1 = grafo_ejemplo. copy()

# Agregar nodos desde un objeto iterable
grafo_ejemplo_1.add_nodes_from(range(9, 14))

# Agregar varios enlaces desde un objeto iterable
grafo_ejemplo_1.add_edges_from([(0, 9),(5, 9),
 (11,12),(12,13),(13,11),(4,12),(0,9) ])

# Eliminar varios enlaces
grafo_ejemplo_1.remove_edges_from([(1, 2), (1, 3), (1, 4),(1, 5)])

# Mostrar el dibujo del grafo
plt.figure(figsize=(5, 5))
nx.draw(grafo_ejemplo_1, with_labels=True, font_size=12)
plt.show()

# Crear la matriz de adyacencia del grafo
matrix_adj_ejemplo_1 = nx.adjacency_matrix(grafo_ejemplo_1)

# Mostrar la matriz de adyacencia
print("Este es el contenido de la matriz de adyacencia:\n",
      matrix_adj_ejemplo_1.todense(), "\n")
```

## Análisis a nivel de nodos

### Grado de un nodo

```python
# Grado de un nodo
print("Este es grado del nodo 2:\n",grafo_ejemplo_1.degree(2), "\n")

# Mostrar el grado de los nodos como lista
print("Estos son los grados de los nodos:\n",nx.degree(grafo_ejemplo_1), "\n")

# Mostrar los grados de los nodos
for nodo in nx.degree(grafo_ejemplo_1):
  print("Nodo:", nodo[0], "  Grado:", nodo[1])
```

### Distancia entre dos nodos

```python
# Mostrar la distancia entre dos nodos
print("Estos son las distancias entre los nodos:\n",
      nx.shortest_path(grafo_ejemplo), "\n")

# Mostrar la distancia entre los nodos
for i in grafo_ejemplo_1:
  for j in grafo_ejemplo_1:
    if i != j:
      try:
        print("La distancia entre", i, "y", j, "es:",
              nx.shortest_path(grafo_ejemplo_1, i, j))
      except:
        pass
```

### Excentricidad de un nodo

Máxima distancia de un nodo a todos los demás nodos del grafo

```python
# Mostrar la excentricidad  de los nodos como diccionario
print("Esta es la excentricidad de los nodos:\n",
      nx.eccentricity(grafo_ejemplo), "\n")

# Mostrar la excentricidad  de los nodos como diccionario
print("Esta es la excentricidad de los nodos:\n",
      nx.eccentricity(grafo_ejemplo_1), "\n")
```

### Grado de centralidad (degree centrality)

El grado de centralidad (degree centrality) es una medida que indica el porcentaje de nodos a los que está conectado un nodo.

```python
# Mostrar el grado de centralidad de los nodos como diccionario
print("Este es el grado de centralidad de los nodos:\n",
      nx.degree_centrality(grafo_ejemplo_1), "\n")

# Mostrar el grado de centralidad de los nodos
for clave, valor in nx.degree_centrality(grafo_ejemplo_1).items():
  print("Nodo:", clave, "  Grado de centralidad:","%.5f"% valor)
```

### Centralidad de cercanía (closeness centrality)

La centralidad de cercanía (closeness centrality) es una medida de centralidad de un nodo que indica que tan cerca está de los otros nodos. Se calcula como la suma de las distancias geodésicas de un nodo a todos los otros nodos de la red.

```python
# Mostrar la centralidad de los nodos como diccionario
print("Esta es la centralidad de los nodos:\n",
      nx.closeness_centrality(grafo_ejemplo_1), "\n")

# Mostrar la centralidad de los nodos
for clave, valor in nx.closeness_centrality(grafo_ejemplo_1).items():
  print("Nodo:", clave, "  Centralidad:","%.5f"% valor)
```

### Centralidad de intermediación (betweenness centrality)

La centralidad de intermediación (betweenness centrality) es la medida del número de veces que hay que pasar por un nodo para alcanzar a otro.

```python
# Mostrar la centralidad de intermediación de los nodos como diccionario
print("Esta es la centralidad de intermediación de los nodos:\n",
      nx.betweenness_centrality(grafo_ejemplo_1), "\n")

# Mostrar la centralidad de intermediación de los nodos
for clave, valor in nx.betweenness_centrality(grafo_ejemplo_1).items():
  print("Nodo:", clave, "  Centralidad de intermediación:","%.5f"% valor)
```

### Centralidad del vector propio (eigenvector centrality)

La centralidad del vector propio (eigenvector centrality) es una medida que indica la centralidad de los nodos que preceden a un nodo; es decir, la cercanía a los nodos de mayor influencia

```python
# Mostrar la centralidad de vector propio de los nodos como diccionario
print("Esta es la centralidad de vector propio de los nodos:\n",
      nx.eigenvector_centrality(grafo_ejemplo_1), "\n")

# Mostrar la centralidad de vector propio de los nodos
for clave, valor in nx.eigenvector_centrality(grafo_ejemplo_1).items():
  print("Nodo:", clave, "  Centralidad de vector propio:","%.5f"% valor)
```

### Vecinos (neighbors)



```python
# Mostrar los vecinos y los no vecinos de un nodo
for i in grafo_ejemplo_1:
  print("Los vecinos de", i, "son:\n",
        list(nx.all_neighbors(grafo_ejemplo_1, i)))
  print("Los no vecinos de", i, "son:\n",
        list(nx.non_neighbors(grafo_ejemplo_1, i)))

# Mostrar los vecinos compartidos entre dos nodos
for i in grafo_ejemplo_1:
  for j in grafo_ejemplo_1:
    if i != j:
      if len(list(nx.common_neighbors(grafo_ejemplo_1, i, j))) != 0:
        print("Los vecinos compartidos de", i, "con", j, "son:\n",
              list(nx.common_neighbors(grafo_ejemplo_1, i, j)))
```

### Puentes (bridges)

En un grafo se considera un puente a un enlace que si se elimina, el número de componentes conectadas se incrementa.

```python
# Mostrar los puentes de un grafo
print("Los puentes del grafo son:\n", list(nx.bridges(grafo_ejemplo_1)))
```

### Coeficiente de agrupamiento (clustering coefficient)

El coeficiente de agrupamiento (clustering coefficient) indica el porcentaje de triadas existentes respecto al total de triadas posibles en un nodo.

```python
# Mostrar el coeficiente de agrupamiento de los nodos como diccionario
print("Esta es el coeficiente de agrupamiento de los nodos:\n",
      nx.clustering(grafo_ejemplo_1), "\n")

# Mostrar el coeficiente de agrupamiento de los nodos
for clave, valor in nx.clustering(grafo_ejemplo_1).items():
  print("Nodo:", clave, "  Coeficiente de agrupamiento:","%.5f"% valor)
```
